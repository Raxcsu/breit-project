```python
data_academico.nivel_de_logro.value_counts()
```

```python
# Custom function to calculate mode
def calculate_mode(series):
    mode = series.mode()
    if not mode.empty:
        return mode.iloc[0]  # Return the first mode if there are multiple
    return np.nan  # Return NaN if no mode exists
```

```python
# Pivot the table
# Convert 'nivel_de_logro' column to numeric, coercing errors to NaN
#data_qm['nivel_de_logro'] = pd.to_numeric(data_qm['puntaje'], errors='coerce')
data_academico = data_academico.pivot_table(
    index = ['periodo', 'id_part_grado', 'participante', 'ruralidad'],  # Group by participant and period
    columns ='curso',       # Use 'habilidad' as new columns
    values ='nivel_de_logro',                     # Use 'puntaje' as values
    aggfunc = calculate_mode                         # Use the custom mode function
).reset_index()  # Reset index to flatten the DataFrame

# Rename columns to remove the multi-index
data_academico.columns.name = None

data_academico.head()
```

### ¿Que significado tiene la columna Promedio?

```python
def check_dist(data_academico):
    data = data_academico.copy()
    for u_val in data.nivel_de_logro.unique():
        min_ = data.loc[data.nivel_de_logro == u_val].promedio.min()
        max_ = data.loc[data.nivel_de_logro == u_val].promedio.max()
        mean_ = data.loc[data.nivel_de_logro == u_val].promedio.mean()
        print(f'{u_val}: min = {min_}, max = {max_}, mean = {mean_}')
```

```python
# Step 1: Replace categorical values with numeric values
replace_dict = {
    'En inicio': 1,
    'En proceso': 2,
    'Satisfactorio': 3
}

data_academico['Comunicación'] = data_academico['Comunicación'].replace(replace_dict).astype(float)
data_academico['Matemática'] = data_academico['Matemática'].replace(replace_dict).astype(float)



# Step 3: Group by id_part_grado and periodo
grouped = data_academico.groupby(['id_part_grado', 'periodo'])

# Step 4: Calculate the mean of Comunicación and Matemática for each row within the group
data_academico['nivel_de_logro_aca'] = grouped.apply(lambda x: x[['Comunicación', 'Matemática']].mean(axis=1), include_groups=False).reset_index(level=[0, 1], drop=True)

# Step 5: Create new columns indicating whether Comunicación and Matemática have values
data_academico['Comunicación'] = data_academico['Comunicación'].notna().astype(int)
data_academico['Matemática'] = data_academico['Matemática'].notna().astype(int)

# Step 2: Replace nulls with 0 in Comunicación and Matemática
data_academico['Comunicación'] = data_academico['Comunicación'].fillna(0)
data_academico['Matemática'] = data_academico['Matemática'].fillna(0)
```
