```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = '../data/sociemocional.csv'
df = pd.read_csv(file)
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.dropna()  # Elimina todas las filas con NaNs
df = df.drop_duplicates()
```

```python
df.info()
```

```python
df
```

```python
# Contar la cantidad de estudiantes en cada nivel de logro
niveles_de_logro = df['Nivel de logro'].value_counts()

# Crear gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(niveles_de_logro.index, niveles_de_logro.values)
plt.title('Distribución de Estudiantes por Nivel de Logro', fontsize=14)
plt.xlabel('Nivel de Logro', fontsize=12)
plt.ylabel('Cantidad de Estudiantes', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Contar la cantidad de estudiantes en cada grado
distribucion_por_grado = df['Grado'].value_counts()

# Crear un orden específico para los grados
orden_grados = ['Pri-1ero', 'Pri-2ndo', 'Pri-3ero', 'Pri-4rto', 'Pri-5nto', 'Pri-6xto', 
                'Sec-1ero', 'Sec-2ndo', 'Sec-3ero', 'Sec-4rto', 'Sec-5nto']

# Reordenar los datos según el orden especificado
distribucion_por_grado = distribucion_por_grado.reindex(orden_grados)

# Crear gráfico de barras para la distribución por grado con el nuevo orden
plt.figure(figsize=(10, 6))
plt.bar(distribucion_por_grado.index, distribucion_por_grado.values) # , color='skyblue'
plt.title('Distribución de Estudiantes por Grado', fontsize=14)
plt.xlabel('Grado', fontsize=12)
plt.ylabel('Cantidad de Estudiantes', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Crear una tabla que cuente los niveles de logro por región educativa
niveles_logro_por_region = df.groupby(['Region_IE', 'Nivel de logro']).size().unstack().fillna(0)

# Crear un gráfico de barras apiladas para comparar los niveles de logro entre regiones
niveles_logro_por_region.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Comparación del Nivel de Logro entre Regiones Educativas', fontsize=14)
plt.xlabel('Región Educativa', fontsize=12)
plt.ylabel('Cantidad de Estudiantes', fontsize=12)
plt.xticks(rotation=90)
plt.legend(title='Nivel de Logro')
plt.show()
```

```python
# Crear una tabla que cuente los niveles de logro por curso
niveles_logro_por_curso = df.groupby(['Curso', 'Nivel de logro']).size().unstack().fillna(0)

# Crear un gráfico de barras apiladas para comparar los niveles de logro entre cursos
niveles_logro_por_curso.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Comparación del Nivel de Logro por Curso', fontsize=14)
plt.xlabel('Curso', fontsize=12)
plt.ylabel('Cantidad de Estudiantes', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Nivel de Logro')
plt.show()
```
